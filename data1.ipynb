```python
def calculate_values(x, y):
    """
    This function performs multiple calculations
    using two input parameters x and y
    """
    
    # Addition
    addition = x + y
    
    # Subtraction
    subtraction = x - y
    
    # Multiplication
    multiplication = x * y
    
    # Division (handle division by zero)
    if y != 0:
        division = x / y
    else:
        division = "Cannot divide by zero"
    
    # Return all results
    return {
        "addition": addition,
        "subtraction": subtraction,
        "multiplication": multiplication,
        "division": division
    }
```
